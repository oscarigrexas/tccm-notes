### Heitler-London

### Pauling
How to improve upon the HL wave function?

Covalent + ionic superposition:

$$\Psi_\textit{VB} = c_1(|a\bar{b}| + |b\bar{a}|) + c_2(|a\bar{a}| + |b\bar{b}|)$$

Commonly refered to as the **quantum dressing of Lewis' model**.

### VB vs MO
What's the difference between Heitler-London and Hartree-Fock then?

HF's wave function always implies a 50% covalent 50% ionic character, which while it's not that bad around the equilibrium distance, messes up the dissociation curve.

The optimal mix is actually around 80% and 20% ionic.

### The 3 main ingredients of VB theory
- Localized orbitals
- Active/inactive separation
- Basis of structures

### Localized orbitals

Let's say we want to construct a VB wave function which corresponds to Lewis's picture, which orbitals do we use?
We may create hybrid orbitals by linearly transforming atomic orbitals.

Electrons occupy localized orbitals (atomic, hybrid...).

A bond is defined as 2 singlet-coupled electrons in 2 localized orbitals (+ minor ionics).


### Active/inactive separation
Active treated at the VB level, inactive treated at the MO level.

### Basis of structures
All possible structures are generated by distributing active electrons with a given spin coupling.

Some structures are excluded by symmetry (state-dependent)

Select a complete set of non-redundant structures (Rumer's basis)

**Rumer's rules for a covalent n-electron/n-orbital system**
1. Put the orbitals around an imaginary circle (even if the molecule isn't a ring)
2. Generate all possible VB structure not displaying crossing bonds

**Rumer's rules for n-center/n-electron ionic structures**
1. Choose a distribution of charges
2. Apply Rumer's rules on the rest of the system
3. Choose another center for the lone pair
4. Repeat!

**Rumer's rules for radicals**
1. Add a ficticious center with one lectron
2. Treat the system as a singlet

### Consequence of ingredient 3: the concept of resonance and mesomery

## Part 2: Ab initio Valence Bond

### VB methods including correlation

### The VBSCF method
Basically a MCSCF method with nonorthogonal orbitals.

Same set of orbitals for all VB structures, optimized for a mean situation.

It would be better if each structure had its own specific set of orbitals optimized for each situation.

### The BOVB (Breathing-Orbital Valence Bond) method
Same number of structures as VBSCF but different orbitals for the different structures.
All wave function parameters (structure coeffcients and orbital coefficient sets) are optimized simultaneously.
